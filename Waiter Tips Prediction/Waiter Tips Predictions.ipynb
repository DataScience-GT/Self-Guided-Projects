Map the sex, smoker, day, and time columns to numerical values.

```python
# Write your code here.
```

<details>

        data["sex"] = data["sex"].map({"Female": 0, "Male": 1})
        data["smoker"] = data["smoker"].map({"No": 0, "Yes": 1})
